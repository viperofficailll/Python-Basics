```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.histplot(data=tips, x="total_bill")
plt.show()
```

```python
sns.histplot(data=tips, x="total_bill", kde=True)
plt.show()
```

```python
sns.histplot(data=tips, x="total_bill", bins=30, kde=True)
plt.show()
```

```python
sns.kdeplot(data=tips, x="total_bill", fill=True)
plt.show()
```

```python
sns.histplot(data=tips, x="total_bill", hue="sex", kde=True)
plt.show()
```

```python
sns.displot(data=tips, x="total_bill", kind="hist", kde=True, height=4, aspect=1.5)
plt.show()
```
