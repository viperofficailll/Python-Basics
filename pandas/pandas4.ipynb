## Pivot tables

```python
import pandas as pd
```

```python


sales_data = {
    "date": pd.date_range(start="2025-01-01", periods=20, freq="15D"),
    "product": [
        "A", "B", "C", "D", "A",
        "B", "C", "D", "A", "B",
        "C", "D", "A", "B", "C",
        "D", "A", "B", "C", "D"
    ],
    "region": [
        "North", "South", "East", "West", "North",
        "East", "South", "West", "North", "South",
        "East", "West", "North", "South", "East",
        "West", "North", "South", "East", "West"
    ],
    "sales": [
        450, 720, 380, 910, 650,
        540, 860, 430, 790, 610,
        500, 980, 340, 890, 760,
        680, 520, 940, 470, 830
    ],
    "units": [
        45, 72, 38, 91, 65,
        54, 86, 43, 79, 61,
        50, 98, 34, 89, 76,
        68, 52, 94, 47, 83
    ],
    "rep": [
        "Alice", "Bob", "Charlie", "David", "Emma",
        "Frank", "Grace", "Henry", "Alice", "Bob",
        "Charlie", "David", "Emma", "Frank", "Grace",
        "Henry", "Alice", "Bob", "Charlie", "David"
    ],
    "month": [
        "Jan", "Jan", "Jan", "Feb", "Feb",
        "Feb", "Mar", "Mar", "Mar", "Apr",
        "Apr", "Apr", "May", "May", "May",
        "Jun", "Jun", "Jun", "Jul", "Jul"
    ],
    "quarter": [
        "Q1", "Q1", "Q1", "Q1", "Q1",
        "Q1", "Q1", "Q1", "Q1", "Q2",
        "Q2", "Q2", "Q2", "Q2", "Q2",
        "Q2", "Q2", "Q2", "Q3", "Q3"
    ]
}

df = pd.DataFrame(sales_data)

print(df)
```

```python
pd.pivot_table(df,"sales",index="region" ,columns="product", aggfunc="max")
```

```python
pd.pivot_table(df,"sales",index="region" ,columns="product", aggfunc="min")
```
