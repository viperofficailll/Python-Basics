```python
import pandas as pd 
import numpy as np 
```

## Finding Missing Data

```python
data = {
    "A": [1, 2, np.nan, 4, 5],
    "B": [np.nan, 2, 3, 4, 5],
    "C": [1, 2, 3, np.nan, np.nan],
    "D": [1, np.nan, np.nan, np.nan, 5],
}
df1 =pd.DataFrame(data)
print(df1)
```

```python
df1.isna()
```

```python
df1.isna().sum()
```

```python
df1.isna().any()
```

## Removing Missing data

```python
df1.dropna(thresh=3)
```

## Filling missing value

```python
df1.fillna(0)
```

```python
values ={"A":5 ,"B":10,"C":100,"D":400}
df1.fillna(value=values)
```

```python
df1.fillna(df1.mean())
```

## Merging joining and concatination

```python
## Merging two data frames

employee_info = {
    "employee_id": [101, 102, 103, 104, 105],
    "name": ["Sandesh", "Aarav", "Priya", "Rohan", "Sita"],
    "department": ["Engineering", "HR", "Finance", "Marketing", "Sales"]
}

employee_compensation = {
    "employee_id": [101, 102, 103, 106, 107],
    "salary": [75000, 60000, 68000, 72000, 65000],
    "bonus": [5000, 3000, 4500, 4000, 3500]
}

df_info = pd.DataFrame(employee_info)
df_compensation = pd.DataFrame(employee_compensation)
pd.merge(df_info,df_compensation ,on="employee_id", how="inner")
```

```python
pd.merge(df_info,df_compensation ,on="employee_id", how="outer")
```

```python
pd.merge(df_info,df_compensation ,on="employee_id", how="left")
```

```python
pd.merge(df_info,df_compensation ,on="employee_id", how="right")
```
