## Pandas Series

A series is a one dimensional labelled array capable of holding any data type. the axis labels are collectively called the index .

## Creating series

```python
labels =[ 'a' ,'b','c']
my_list = [10,20,30]
my_arr =np.array([10,20,30])
my_dict = {"a":10,"b":20,"c":30}
pd.Series(my_list ,index=labels)
pd.Series(my_dict)
```

## Data Frame

```python
people_dict = {
    "name": ["Sandesh", "Aarav", "Priya", "Rohan"],
    "age": [24, 28, 26, 30],
    "city": ["Pokhara", "Kathmandu", "Lalitpur", "Biratnagar"],
    "salary": [65000, 80000, 72000, 90000]
}
df1 =pd.DataFrame(people_dict)
```

```python
people_list = [
    ["Sandesh", 24, "Pokhara", 65000],
    ["Aarav", 28, "Kathmandu", 80000],
    ["Priya", 26, "Lalitpur", 72000],
    ["Rohan", 30, "Biratnagar", 90000]
]
columns= ["name" ,"age" ,"city" ,"salary"]
df2 =pd.DataFrame(people_list , columns=columns)
```
