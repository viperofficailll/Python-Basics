## Groupby and aggregation

```python
import pandas as pd

store_sales = {
    "category": ["A", "B", "A", "B", "A", "B", "A", "B"],
    "store": ["S1", "S2", "S1", "S2", "S1", "S2", "S1", "S2"],
    "sales": [120, 340, 280, 450, 190, 390, 260, 480],
    "quantity": [12, 25, 18, 42, 15, 37, 22, 49],
    "date": pd.date_range(start="2025-01-01", periods=8),
}

df1 = pd.DataFrame(store_sales)
df1
```

## GroupBY

```python
category_wise_data = df1.groupby("category")["sales"].sum()
category_wise_data
```

```python
store_wise_data = df1.groupby("store")["sales"].sum()
store_wise_data
```

```python
combined_data = df1.groupby(["category","store"])["sales"].sum()
combined_data
```
