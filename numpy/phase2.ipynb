## Numpy array operations

```python
import numpy as np
```

```python
arr = np.arange(10)
print(arr)
print("Basic slicing",arr[2:7])
print("With step ",arr[1:8:2])
print("Nagative indexing" ,arr[-3])
```

```python
arr_2d =np.array([np.arange(5),np.arange(5),np.arange(5)])
print(arr_2d)
print("specific element ", arr_2d[1,2])
print("Entire row", arr_2d[1])
print("Entire coloum",arr_2d[:,1])
```

## Sorting

```python
unsorted = np.array([np.random.randint(1,101,size=10)])
print(unsorted)
print("Sorted array ",np.sort(unsorted))
```
