## Numpy array operations

```python
import numpy as np
```

```python
arr = np.arange(10)
print(arr)
print("Basic slicing",arr[2:7])
print("With step ",arr[1:8:2])
print("Nagative indexing" ,arr[-3])
```

```python
arr_2d =np.array([np.arange(5),np.arange(5),np.arange(5)])
print(arr_2d)
print("specific element ", arr_2d[1,2])
print("Entire row", arr_2d[1])
print("Entire coloum",arr_2d[:,1])
```

## Sorting

```python
unsorted = np.array([np.random.randint(1,101,size=10)])
print(unsorted)
print("Sorted array ",np.sort(unsorted))
```

## Filters

```python
numbers = np.arange(10)
even_numbers = numbers[numbers % 2 == 0]
print(even_numbers)
```

## Indexing

```python
indices = [0,2,4]
print(numbers[indices])
```

## Where

```python
where_results = np.where(numbers<5)
print("less than 5 ",numbers[where_results])
```

## Conditionals

```python
conditionArray =np.where(numbers>5 ,"true" , "fales")
print(conditionArray)
```

## Adding and removing data

```python
arr1 =np.arange(0,3,1)
arr2= np.arange(3,6,1)
print(arr1)
print(arr2)
combined = np.concatenate((arr1,arr2))
print(combined)
```

## Compatibility test

```python
a = np.array([1,2,3])
b = np.array([1,2,3,4])
c = np.array([1,2,3])
print("Compatibililty",a.shape == b.shape)
```

```python
original = np.array([[1,2,],[2,3]])
new_row =np.array([[5,6]])
with_new_row = np.vstack((original,new_row))
print(original)
print(with_new_row)

new_col = np.array([[7],[8]])
with_new_col = np.hstack((original,new_col))
print(with_new_col)
```

## Deleting

```python
arr = np.array([1,2,3,4,5])
deleted = np.delete(arr, 2)
print(deleted)
```
