```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure : [restaurant_id , 2021 ,2022 ,2023 ,2024]


sales_data = np.array(
    [
        [1, 150000, 180000, 220000, 250000],  # Paradise biryani
        [2, 120000, 140000, 160000, 190000],  # Beijing bites
        [3, 200000, 230000, 260000, 300000],  # pizza hub
        [4, 180000, 210000, 240000, 270000],  # Burger point
        [5, 160000, 185000, 205000, 230000],  # chill point
    ]
)
print("====Zomato sales analysis==== \n")
print("sales data shape ", sales_data.shape)
print(" \n sample data for first 3 restaurant \n", sales_data[0:3])
```

```python
# Total sales per year
yearly_total = np.sum(sales_data[:, 1:], axis=0)
print(yearly_total)
```

```python
# Minimum sales per restaurant
min_sales = np.min(sales_data[:, 1:], axis=1)
print("Min sales  per restaurant \n", min_sales)
```

```python
# Maximum sales in a year by a restaurant
max_sales = np.max(sales_data[:, 1:], axis=0)
print("max sales in a  year \n", max_sales)
```

```python
# Average sales per restaurant
avg_sales = np.mean(sales_data[:, 1:], axis=1)
print("Average sales per year", avg_sales)
```

```python
cumsum = np.cumsum(sales_data[:, 1:], axis=1)
print(cumsum)
plt.figure(figsize=(10, 5))
plt.plot(np.mean(cumsum, axis=0))
plt.title("Average cumulative sales")
plt.xlabel("years")
plt.ylabel("amount")
plt.grid()
plt.show()
```

```python
years = [2021, 2022, 2023, 2024]
restaurant_names = [
    "Paradise Biryani",
    "Beijing Bites",
    "Pizza Hub",
    "Burger Point",
    "Chill Point",
]

plt.figure(figsize=(10,6))
for i in range (len(sales_data)):
    plt.plot(years,sales_data[i,1:] ,marker ="o" ,label =restaurant_names[i])
plt.title('Restaurant sales per year')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.grid(True)
plt.legend()
plt.show()
```
