## Creating arrays from scratch

```python
zeros = np.zeros((3, 4))
print(zeros)
ones = np.ones((2, 3))
print("array of ones \n",ones)

sevens = np.full((2,2),7)
print(sevens)
```

## Random in numpy

```python
random = np.random.random((2, 3))
print("random array \n", random)
sequence = np.arange(0, 10, 2)
print("sequence array : \n", sequence)
```

## Vector , Matrix and Tensor

```python
## A vector is a single row or a single column of numbers. 
# It has exactly one dimension (Axis 0). 
# You can think of it like a straight line of data points or a simple shopping list.
vector = np.array([1, 2, 3])
print("vector \n", vector)
```

```python
## A matrix is a grid of numbers arranged in rows and columns.
## It has two dimensions (Axis 0 for rows, Axis 1 for columns).
## Think of it like an Excel spreadsheet or a digital grayscale image (where each pixel is a row/column coordinate).
mat = np.array([[1, 2], [3, 4]])
print("matrix \n", mat)
```

```python
## A tensor is the umbrella term for any multi-dimensional array.
# However, in practice, developers usually say "tensor"
# when talking about arrays with three or more dimensions.

tensor = np.array([[1, 2], [3, 4], [5, 6], [7, 8]])
print("tensor \n", tensor)
```

## Array Properties

```python
arr = np.array([[1, 2, 3], [4, 5, 6]])
print("shape \n", arr.shape)
print("dimension \n", arr.ndim)
print("size \n", arr.size)
print("Data types \n", arr.dtype)
```

## Array reshaping

```python
arr = np.arange(12)
print("original array \n", arr)
reshaped = arr.reshape((3, 4))
print("reshaped array \n", reshaped)
flattend = reshaped.flatten()
print("flattend array \n", flattend)
raveled = reshaped.ravel()
print(" raveled array \n", raveled)
transposed = arr.T
print("transposed array \n",transposed)
```
